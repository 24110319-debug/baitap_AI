```python
import random
import copy
```

```python
goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
#state 
state = []
print("Nhập trạng thái bàn cờ 3x3")
for i in range(3):

    row = list(map(int, input().split()))

    state.append(row)
# MODEL
model = []

# ACTION TRƯỚC ĐÓ
previous_action = None

# RULES
rules = ["UP", "DOWN", "LEFT", "RIGHT"]

```

```python
# HIỂN THỊ
def print_board(board):

    for row in board:

        print(row)

    print()
```

```python
# TÌM Ô TRỐNG
def find_blank(board):

    for i in range(3):

        for j in range(3):

            if board[i][j] == 0:

                return i, j
```

```python
def board_to_string(board):

    return str(board)
```

```python
def move(board, action):

    x, y = find_blank(board)

    new_board = copy.deepcopy(board)

    if action == "UP" and x > 0:

        new_board[x][y], new_board[x - 1][y] = \
        new_board[x - 1][y], new_board[x][y]

    elif action == "DOWN" and x < 2:

        new_board[x][y], new_board[x + 1][y] = \
        new_board[x + 1][y], new_board[x][y]

    elif action == "LEFT" and y > 0:
        new_board[x][y], new_board[x][y - 1] = \
        new_board[x][y - 1], new_board[x][y]

    elif action == "RIGHT" and y < 2:

        new_board[x][y], new_board[x][y + 1] = \
        new_board[x][y + 1], new_board[x][y]

    return new_board
```

```python
visited = []

def reflex_agent(board):

    actions = ["UP", 
               "DOWN", 
               "LEFT", 
               "RIGHT"]

    random.shuffle(actions)

    for action in actions:

        new_board = move(board, action)

        # tránh trạng thái lặp trạng thái
        if board_to_string(new_board) not in visited:

            return action, new_board

    return "STOP", board
```

```python
step = 1

print("\nINITIAL STATE")
print_board(state)

while True:

    # kiểm tra đích
    if state == goal_state:

        print("PUZZLE SOLVED")

        break
    visited.append(board_to_string(state))

    action, new_state = reflex_agent(state)

    if action == "STOP":

        print("NO MORE MOVE")

        break

    print("STEP =", step)
    print("ACTION =", action)

    state = new_state

    print_board(state)

    step += 1
```
