```python
import random
```

```python
# MÔI TRƯỜNG
# 0 = CLEAN (sạch)
# 1 = DIRTY (bẩn)

environment = [
    [1, 0, 1],
    [0, 1, 0],
    [1, 0, 1]
]

# KÍCH THƯỚC MÔI TRƯỜNG

n = 3


# STATE

state = {
    "x": 0,
    "y": 0
}

# MODEL:Lưu các vị trí đã đi

model = []

# RULES

rules = ["UP", "DOWN", "LEFT", "RIGHT"]


# ACTION TRƯỚC ĐÓ

action = None

```

```python
# HIỂN THỊ MÔI TRƯỜNG

def print_environment():

    for i in range(n):

        for j in range(n):

            if i == state["x"] and j == state["y"]:

                print("R", end=" ")

            else:

                print(environment[i][j], end=" ")

        print()
    print()
```

```python
# KIỂM TRA SẠCH HẾT

def all_clean():

    for row in environment:

        for cell in row:

            if cell == 1:

                return False

    return True
```

```python
# UPDATE STATE

def update_state(state, action):

    x = state["x"]
    y = state["y"]

    if action == "UP" and x > 0:

        x -= 1

    elif action == "DOWN" and x < n - 1:

        x += 1

    elif action == "LEFT" and y > 0:
         
        y -= 1
    elif action == "RIGHT" and y < n - 1:

        y += 1

    return {
        "x": x,
        "y": y
    }
```

```python
# RULE MATCH

def rule_match(state, rules, model):

    actions = []

    x = state["x"]
    y = state["y"]

    # KIỂM TRA HƯỚNG HỢP LỆ

    if x > 0:

        actions.append("UP")

    if x < n - 1:
         actions.append("DOWN")

    if y > 0:

        actions.append("LEFT")

    if y < n - 1:

        actions.append("RIGHT")
        
    # RANDOM ACTION

    random.shuffle(actions)

    for act in actions:

        new_state = update_state(state, act)

        # tránh lặp vị trí
        if str(new_state) not in model:

            return act

    return "STOP"
```

```python
# MODEL-BASED REFLEX AGENT

def model_based_reflex_agent():

    global state
    global action
    global model

    x = state["x"]
    y = state["y"]

    percept = environment[x][y]

    # IF DIRTY -> SUCK

    if percept == 1:
        
        print("DIRTY -> SUCK")

        environment[x][y] = 0

    else:

        print("CLEAN")
        
 # LƯU STATE VÀO MODEL

    model.append(str(state))

    # CHỌN ACTION

    action = rule_match(state, rules, model)

    return action
```

```python
# CHẠY AGENT

step = 1
max_step = 30

print("INITIAL ENVIRONMENT")
print_environment()

while step <= max_step:

    if all_clean():

        print("ALL CLEAN -> STOP")

        break

    print("STEP =", step)

    action = model_based_reflex_agent()

    print("ACTION =", action)

    if action == "STOP":

        print("NO MOVE")

        break
    
 # UPDATE STATE

    state = update_state(state, action)

    print_environment()

    step += 1


if step > max_step:

    print("STOP - TOO MANY STEPS")
    
```
